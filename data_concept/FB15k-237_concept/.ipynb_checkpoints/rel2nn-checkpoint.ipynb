```python
rel_id = {}
id_rel = {}
f = open("./relations.dict", 'r')
for line in f:
    id, rel = line.strip().split("\t")
    rel_id[rel] = int(id)
    id_rel[int(id)] = rel
f.close()
```

```python
rel2nn = {}
f = open("./1-1.txt", 'r')
for line in f:
    rel = line.strip().split("\t")[1]
    if rel not in rel2nn:
        rel2nn[rel_id[rel]] = 0
f.close()
```

```python
f = open("./1-n.txt", 'r')
for line in f:
    rel = line.strip().split("\t")[1]
    if rel not in rel2nn:
        rel2nn[rel_id[rel]] = 1
f.close()
```

```python
f = open("./n-1.txt", 'r')
for line in f:
    rel = line.strip().split("\t")[1]
    if rel not in rel2nn:
        rel2nn[rel_id[rel]] = 2
f.close()
```

```python
f = open("./n-n.txt", 'r')
for line in f:
    rel = line.strip().split("\t")[1]
    if rel not in rel2nn:
        rel2nn[rel_id[rel]] = 3
f.close()
```
