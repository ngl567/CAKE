```python
dom_id = {}
id_dom = {}
f = open("./domain2id.txt", 'r')
for line in f:
    dom, id = line.strip().split("\t")
    dom_id[dom] = int(id)
    id_dom[int(id)] = dom
f.close()
```

```python
ent_id = {}
id_ent = {}
fin = open("./entity2id.txt", 'r')
f = open("./entities.dict", 'w')
for line in fin:
    ent, id = line.strip().split("\t")
    ent_id[ent] = int(id)
    id_ent[int(id)] = ent
    f.write(id + "\t" + ent + "\n")
f.close()
fin.close()
```

```python
ent_dom = {}
f = open("./entity2domain.txt", 'r')
for line in f:
    s = line.strip().split("\t")
    ent = s[0]
    ent_dom[ent_id[ent]] = []
    conc_n = int(s[1])
    for i in range(2, len(s)):
        ent_dom[ent_id[ent]].append(dom_id[s[i]])
f.close()
```

```python
len(dom_id)
```

```python
import json
with open('ent_dom.json', 'w', encoding='utf-8') as f:
    json.dump(ent_dom, f)
```

```python
dom_ent = {}
for ent, doms in ent_dom.items():
    for dom in doms:
        if dom not in dom_ent:
            dom_ent[dom] = []
        dom_ent[dom].append(ent)
```

```python
rel_id = {}
id_rel = {}
fin = open("./relation2id.txt", 'r')
f = open("./relations.dict", 'w')
for line in fin:
    rel, id = line.strip().split("\t")
    rel_id[rel] = int(id)
    id_rel[int(id)] = rel
    f.write(id + "\t" + rel + "\n")
f.close()
fin.close()
```
