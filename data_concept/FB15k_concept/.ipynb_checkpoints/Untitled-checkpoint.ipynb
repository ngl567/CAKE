```python
dom_id = {}
f = open("./domain2id.txt", 'r')
for line in f:
    dom, id = line.strip().split("\t")
    dom_id[dom] = int(id)
f.close()
```

```python
ent_id = {}
f = open("./entities.dict", 'r')
for line in f:
    id, ent = line.strip().split("\t")
    ent_id[ent] = int(id)
f.close()
```

```python
ent_dom = {}
f = open("./entity2domain.txt", 'r')
for line in f:
    s = line.strip().split("\t")
    ent = s[0]
    ent_dom[ent_id[ent]] = []
    conc_n = int(s[1])
    for i in range(2, len(s)):
        ent_dom[ent_id[ent]].append(dom_id[s[i]])
f.close()
```

```python
len(dom_id)
```

```python
import json
with open('ent_dom.json', 'w', encoding='utf-8') as f:
    json.dump(ent_dom, f)
```

```python
dom_ent = {}
for ent, doms in ent_dom.items():
    for dom in doms:
        if dom not in dom_ent:
            dom_ent[dom] = []
        dom_ent[dom].append(ent)
```

```python
import json
with open('dom_ent.json', 'w', encoding='utf-8') as f:
    json.dump(dom_ent, f)
```

```python
rel_id = {}
f = open("./relations.dict", 'r')
for line in f:
    id, rel = line.strip().split("\t")
    rel_id[rel] = int(id)
f.close()
```
